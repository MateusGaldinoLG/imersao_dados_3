# Estatística e Visualização de dados

Drug discovery = usa muitas drogas para conseguir achar uma que seja útil para o que se quer testar.

Para renomear os dados, é usado o método rename(column = mapa), onde mapa é um dicionário (quase um objeto). Para ele alterar definitivamente, é necessario usar o inplace=True

Para chamar uma variável de uma stream de código para outra, é usado uma @  
Para encontrar o número de tipos únicos, pode ser usada a função lens()  
No Pandas, é possível criar um histograma com **.hist()**  
* Dentro de hist, é possível passar bins='numero' (sendo numero o número de 'caixinhas')  

Com o método describe() do Pandas, é possível saber detalhes importantes sobre cada uma das colunas como média, minimo, máximo, etc  
Slice - .loc[min:max, coluna_min:coluna_max]  
rotação - .T (transforma linhas em colunas e vice versa)

## Seaborn 
Biblioteca de visualização de dados voltada à estatística (possui diversos níveis de abstração para construção de gráfico
import seaborn as sns  
Para plotar um gráfico de barras que conta, é usado o sns.countplot(data=dados, order=['ordem', 'separada', 'por', 'array'])  
O método set_title() coloca um título no gráfico do Seaborn  
boxplot() - plota um gráfico boxplot - gráfico de um único eixo onde é marcado um ponto para cada um dos. O traço vertical indica a mediana dos dados, sendo circulada pelo retângulo azul no gráfico. Os pontos mais afastados do padrão são chamados de *outliers*. O 'miolo', onde a maioria dos valores estão, são representados pelo bloco azul.

## Matplotlib
plt.figure -> configura os detalhes da figura do gráfico  
figsize=(x,y) -> parametro de figure, muda o tamanho da figura para o tamanho x e o y ((x,y) é uma **tupla**)

```python
import pandas as pb
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pb.read_csv(url_dados, compression='zip') 
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados
```

```python
compostos = dados['composto'].value_counts().index[0:5] # pega os 5 primeiros valores [0,5)
compostos 
```

```python
dados.query('composto in @compostos')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set() # setta o design do countplot para padrão

plt.figure(figsize=(8, 6)) 

ax = sns.countplot(x='composto', data=dados.query('composto in @compostos'), order=['cacb2b860', '87d714366', '9f80f3f77', '8b87a7a83', '5628cb3ee'])
ax.set_title("top 5 compostos")
#ax.set(xlabel='composto', ylabel='usos em experimento') -  1 forma de fazer, para os dois
ax.set_ylabel('usos em experimento') # - 2 forma de fazer, para só um
plt.show()
```
